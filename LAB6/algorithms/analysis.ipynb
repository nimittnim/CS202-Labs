```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
parallel_data = pd.read_csv("test_logs/parallel_test_results.csv")
parallel_data
```

```python
# Data
configs = [ "1-1-load", "1-1-no", "1-auto-load", "1-auto-no",
           "auto-1-load", "auto-1-no", "auto-auto-load", "auto-auto-no"]
execution_times = [4.79, 4.86, 33.51, 35.37, 5.10, 5.16, 32.49, 31.64]
# Plot
plt.figure(figsize=(10, 5))
plt.bar(["sequential"],[4.47],color='skyblue')
plt.bar(configs, execution_times, color="red")
plt.xlabel("Configuration (Worker-Thread-Dist Mode)")
plt.ylabel("Execution Time (s)")
plt.title("Execution Time for Different Testing Strategies")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# Data
configs1 = [ "1-1-load", "1-1-no","auto-1-load", "auto-1-no" ]
configs2 = ["1-auto-load", "1-auto-no", "auto-auto-load", "auto-auto-no"]

# Plot
plt.figure(figsize=(10, 5))
plt.bar(configs1,[0,0,0,0],color='skyblue')
plt.bar(configs2, [4,4,4,4], color="red")
plt.xlabel("Configuration (Worker-Thread-Dist Mode)")
plt.ylabel("Number of Failed Tests")
plt.title("Number of failed tests for Different Testing Strategies")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
speedup = np.array(execution_times)/4.47
```

```python
# Data
configs = [ "1-1-load", "1-1-no", "1-auto-load", "1-auto-no",
           "auto-1-load", "auto-1-no", "auto-auto-load", "auto-auto-no"]
# Plot
plt.figure(figsize=(10, 5))
plt.bar(["sequential"],[1],color='skyblue')
plt.bar(configs, speedup, color="red")
plt.xlabel("Configuration (Worker-Thread-Dist Mode)")
plt.ylabel("Speedup")
plt.title("SpeedUp WRT Sequential Testing")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
