```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
lcom_data = pd.read_csv("lcom_out/TypeMetrics.csv")
lcom_data
```

```python
def normalise(ser):
    max_val = np.max(ser)
    min_val = np.min(ser)
    return (ser - min_val) / (max_val - min_val)

lcom_names = ['LCOM1', 'LCOM2', 'LCOM3', 'LCOM4', 'LCOM5']
for lcom_name in lcom_names:
    lcom_data[lcom_name] = normalise(lcom_data[lcom_name])
```

```python
lcom_data
```

```python
lcom_names = ['LCOM1', 'LCOM2', 'LCOM3', 'LCOM4', 'LCOM5']
num = len(lcom_names)

fig, axes = plt.subplots(num, num, figsize=(15, 15))
fig.suptitle("Pairwise Scatter Plots of LCOM Metrics", fontsize=16)

for i in range(num):
    for j in range(num):
        ax = axes[i, j]
        if i == j:
            ax.text(0.5, 0.5, lcom_names[i], fontsize=12,
                    ha='center', va='center')
            ax.set_xticks([])
            ax.set_yticks([])
        else:
            ax.scatter(lcom_data[lcom_names[j]], lcom_data[lcom_names[i]], alpha=0.6)
        
        if i == num - 1:
            ax.set_xlabel(lcom_names[j])
        else:
            ax.set_xticks([])
        
        if j == 0:
            ax.set_ylabel(lcom_names[i])
        else:
            ax.set_yticks([])

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()

```

```python
lcom_data['LCOM_A'] = lcom_data['LCOM1'] + lcom_data['LCOM2'] + lcom_data['LCOM3'] + lcom_data['LCOM4'] + lcom_data['LCOM5'] 
```

```python
lcom_data_sorted = lcom_data.sort_values("LCOM_A", ascending=False)
```
